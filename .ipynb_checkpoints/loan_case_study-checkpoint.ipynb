### Data Cleaning
Data Preparation and Cleaning steps.

```python
#Checking the Missing values.
loan_master.isnull().sum()
```

```python
#Checking missing value percentages of the columns.
round(loan_master.isnull().sum()/len(loan_master.index), 2)*100
```

```python
#Removing columns having missing values over 70%
missing_value_col = loan_master.columns[100*(loan_master.isnull().sum()/len(loan_master.index)) > 60]
print(missing_value_col)
```

```python
#Dropping the missing values from the dataframe.
loan = loan_master.drop(missing_value_col, axis=1)
loan.shape
```

```python
#Checking missing value percentages of the columns.
pd.set_option('display.max_rows', 60)
loan.isnull().sum()
round(loan.isnull().sum()/len(loan.index), 2)*100
```

```python
#Checking the column with 33% missing values.
loan.desc.head()
```

```python
#Desc column is a comments/review column, which might not be very useful for analysis.
#Removing the desc column.
loan = loan.drop('desc', axis=1)
```

```python
#Checking missing value percentages of the columns. 
#All other columns have acceptable % of missing values. We do not intend to build a model, so we can leave a few missng value.
round(loan.isnull().sum()/len(loan.index), 2)*100
```

```python
#Checking missing values in Rows.
#Checking rows having missing values greater than 5.
loan[loan.isnull().sum(axis=1) > 5]
```

```python
#The data is almost clean. We can check the datatype of each column through info() command.
loan.info()
```
